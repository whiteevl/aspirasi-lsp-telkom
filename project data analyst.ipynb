```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## Data Wrangling

### Gathering Data

```python
dataset = pd.read_csv("hour.csv")
dataset = pd.read_csv("day.csv")
print(dataset.head())
```

### Assessing Data

```python
info_dataset = dataset.info()
descriptive_stats = dataset.describe()

print(info_dataset)
print(descriptive_stats)
```

### Cleaning Data

```python
dataset = dataset.dropna()

Q1 = dataset['registered'].quantile(0.25)
Q3 = dataset['registered'].quantile(0.75)
IQR = Q3 - Q1
dataset = dataset[(dataset['registered'] >= Q1 - 1.5 * IQR) & (dataset['registered'] <= Q3 + 1.5 * IQR)]
```

## Exploratory Data Analysis (EDA)

### Explore ...

```python
sns.histplot(dataset['registered'], kde=True)
plt.title("Distribusi Kolom")
plt.show()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
sns.lineplot(x='instant', y='registered', data=dataset[dataset['workingday'] == 1])
plt.title("Tren Penggunaan Sepeda pada Jam Tertentu di Hari Kerja")
plt.show()
```

### Pertanyaan 2:

```python
sns.boxplot(x='workingday', y='registered', data=dataset)
plt.title("Perbedaan Pola Penggunaan Sepeda antara Hari Kerja dan Akhir Pekan")
plt.show()
```
